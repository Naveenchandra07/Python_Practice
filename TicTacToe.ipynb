```python
class TicTacToe:
    
    def __init__(self):

        self.boardSizeInt = 0
        self.board = [[]]
        self.mapPosBoard = {}
    
    def play(self):
        
        boardSizeStr = input("Please provide the size of board: ")
        self.boardSizeInt = self.to_number(boardSizeStr)
        
        if(self.boardSizeInt < 3):
            print("Please provide a numeric value which is greater than or equal to 3 as board size. \nExiting the program!")
        else:
            print()
            print("User 1 will be X and User 2 will be O")
            print()
            rows, cols = (self.boardSizeInt, self.boardSizeInt)
            self.board = [[" * " for i in range(cols)] for j in range(rows)]

            self.map_position_board()

            currPos = 0
            isFirstUser = True
            isBoardFull = False

            while(isBoardFull == False):

                if(self.is_board_full()):
                    isBoardFull = True
                    self.print_board()
                    print("Board is Full! Its a Draw!\n End of Game..")
                    break
                else:
                    self.print_board()
                    currPosStr = input("User {}, please choose the position: ".format("1" if isFirstUser else "2"))
                    currPos = self.to_number(currPosStr)

                    if(currPos <= 0):
                        print("Please provide a numeric value in between 1 and {}".format(totalElements))
                        pass
                    else:
                        if(self.is_pos_occupied(self.mapPosBoard[currPos])):
                            print("This position is already occupied. Please provide a different position")
                            self.print_board()
                        else:
                            rowColList = self.mapPosBoard[currPos]
                            element = " X " if isFirstUser else " O "
                            print()
                            self.board[rowColList[0]][rowColList[1]] = element
                            isFirstUser = False if isFirstUser == True else True
                            
                            if self.validate(currPos) == True:
                                self.print_board()
                                print(element + "is the winner! \n End of Game..")
                                break
    
    def map_position_board(self):
        
        row = 0
        col = 0
        index = 1

        totalElements = self.boardSizeInt * self.boardSizeInt
        for i in range(0, totalElements):
            list_row_col = [row, col]
            self.mapPosBoard[index] = list_row_col
            if(index % self.boardSizeInt == 0):
                row = row + 1
                col = 0
            else:
                col = col + 1
            index = index + 1
    
    def print_board(self):
        for i in range(0, self.boardSizeInt):
            for j in range(0, self.boardSizeInt):
                if(j == self.boardSizeInt - 1):
                    print(self.board[i][j], end="")
                else:
                    print(self.board[i][j], end="|")
            print()
            if(i < self.boardSizeInt - 1):
                for j in range(0, self.boardSizeInt):
                    print(u"\u2015\u2015\u2015", end=" ")    
            print()
        
    def to_number(self, inputVal):
        try:
            return int(inputVal)
        except ValueError:
            return 0
        
    def is_board_full(self):
        isBoardFull = True
        for row in self.board:
            for col in row:
                if col == " * ":
                    isBoardFull = False
                    break
                    
        return isBoardFull
    
    def is_pos_occupied(self, rowColList):
        isPosOccupied = True
        if(self.board[rowColList[0]][rowColList[1]] == " * "):
            isPosOccupied = False
        return isPosOccupied
    
    def validate(self, currPos):
        
        hasWon = False
        
        if(self.validate_diagonal_rule() == True):
            hasWon = True
            
        elif(self.validate_vertical_rule(currPos) == True):
            hasWon = True
            
        elif(self.validate_horizontal_rule(currPos) == True):
            hasWon = True
            
        return hasWon
        
    def validate_diagonal_rule(self):
        
        curPos = 1
        isSame = True
        curList = self.mapPosBoard[curPos]
        curElement = self.board[curList[0]][curList[1]]
        
        while(isSame == True):
            
            nextPos = curPos + self.boardSizeInt + 1
            if(nextPos > (self.boardSizeInt * self.boardSizeInt)):
                break
            
            nextList = self.mapPosBoard[nextPos]
            nextElement = self.board[nextList[0]][nextList[1]]
            
            if(curElement != " * " and nextElement != " * " and curElement == nextElement):
               curPos = nextPos
            else:
                isSame = False
                break
        
        if isSame == False:
            
            isSame = True
            curPos = self.boardSizeInt
            curList = self.mapPosBoard[curPos]
            curElement = self.board[curList[0]][curList[1]]
            
            while(isSame == True):
                nextPos = curPos + self.boardSizeInt - 1
                
                if(nextPos >= (self.boardSizeInt * self.boardSizeInt)):
                    break
                
                nextList = self.mapPosBoard[nextPos]
                nextElement = self.board[nextList[0]][nextList[1]]
                
                if(curElement != " * " and nextElement != " * " and curElement == nextElement):
                   curPos = nextPos
                else:
                    isSame = False
                    break
            
        return isSame
    
    def validate_vertical_rule(self, curPos):
        
        isSame = True
        curPosTemp = curPos
        
        while(curPosTemp > 0):
            if((curPosTemp - self.boardSizeInt) <= 0):
                break
                
            curPosTemp = curPosTemp - self.boardSizeInt
        
        curList = self.mapPosBoard[curPosTemp]
        curElement = self.board[curList[0]][curList[1]]
            
        while(isSame == True):
            
            nextPos = curPosTemp + self.boardSizeInt
            
            if(nextPos > (self.boardSizeInt * self.boardSizeInt)):
                break
                
            nextList = self.mapPosBoard[nextPos]
            nextElement = self.board[nextList[0]][nextList[1]]
            
            if(curElement != " * " and nextElement != " * " and curElement == nextElement):
               curPosTemp = nextPos
            else:
                isSame = False
                break
                
        return isSame
    
    def validate_horizontal_rule(self, currPos):
        
        currPosTemp = 1
        isSame = True
        
        if currPos > self.boardSizeInt:
            currPosTemp = currPos
            while(currPosTemp > 0):
                currPosTemp = currPosTemp - 1
                if(currPosTemp % self.boardSizeInt == 0):
                    break
            currPosTemp = currPosTemp + 1
        
        currList = self.mapPosBoard[currPosTemp]
        element = self.board[currList[0]][currList[1]]
        
        for i in range (currPosTemp, currPosTemp + self.boardSizeInt - 1):
            
            nextPos = i + 1
            
            nextList = self.mapPosBoard[nextPos]
            nextElement = self.board[nextList[0]][nextList[1]]
            
            if(element != " * " and nextElement != " * " and element == nextElement):
               curPosTemp = nextPos
            else:
                isSame = False
                break
        
        return isSame
    
    def test(self):
        
        #Test 3X3 Board
        winningSeries = [[1,5,9,8], [3,5,7,8], [1,4,7,6], [2,5,8,9], [3,6,9,5], [1,2,3,7], [4,5,6,8], [7,8,9,1]]
        drawSeries = [[2,4,5,9], [8,5,1,3]]
        
        self.boardSizeInt = 3
        self.map_position_board()
        
        rows, cols = (3, 3)
        
        print("Testing for Winning Combinations..\n")
        
        for eachSeries in winningSeries:
            
            self.board = [[" * " for i in range(cols)] for j in range(rows)]
            
            for row in range(3):
                for col in range(3):
                    rowColList = [row,col]
                    if (self.mapPosBoard[eachSeries[0]] == rowColList or self.mapPosBoard[eachSeries[1]] == rowColList 
                    or self.mapPosBoard[eachSeries[2]] == rowColList or self.mapPosBoard[eachSeries[3]] == rowColList):
                        self.board[row][col] = " X "
                    else:
                        self.board[row][col] = " O "
            
            if(self.validate(eachSeries[0])):
                print("X has Won!\nTest passed for below input board data\n")
                self.print_board()
            else:
                print("Test Failed! Below is the input data")
                self.print_board()
        
        print("\nTest End!!\n")
        
        print("Testing for Draw Combintions..\n")
        
        for eachSeries in drawSeries:
            
            self.board = [[" * " for i in range(cols)] for j in range(rows)]
            
            for row in range(3):
                for col in range(3):
                    rowColList = [row,col]
                    if (self.mapPosBoard[eachSeries[0]] == rowColList or self.mapPosBoard[eachSeries[1]] == rowColList 
                    or self.mapPosBoard[eachSeries[2]] == rowColList or self.mapPosBoard[eachSeries[3]] == rowColList):
                        self.board[row][col] = " X "
                    else:
                        self.board[row][col] = " O "
            
            if(self.validate(eachSeries[0]) == False):
                print("Ita a Draw!\nTest passed for below input board data\n")
                self.print_board()
            else:
                print("Test Failed! Below is the input data")
                self.print_board()
        
        print("\nTest End!!\n")
```
