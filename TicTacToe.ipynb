```python
class TicTacToe:
    def __init__(self):
        
        boardSizeStr = input("Please provide the size of board: ")
        boardSizeInt = self.to_number(boardSizeStr)
        if(boardSizeInt < 3):
            print("Please provide a numeric value which is greater than or equal to 3 as board size. \nExiting the program!")
        else:
            print()
            rows, cols = (boardSizeInt, boardSizeInt)
            self.board = [[" * " for i in range(cols)] for j in range(rows)]

            map_index_matrix = {}
            row = 0
            col = 0
            index = 1
            totalElements = boardSizeInt*boardSizeInt
            for i in range(0, totalElements):
                list_row_col = [row, col]
                map_index_matrix[index] = list_row_col
                if(index % boardSizeInt == 0):
                    row = row + 1
                    col = 0
                else:
                    col = col + 1
                index = index + 1

            user1 = "X"
            user2 = "O"

            currPos = 0
            isFirstUser = True
            isBoardFull = False

            while(isBoardFull == False):

                if(self.is_board_full(self.board)):
                    isBoardFull = True
                    self.print_board(self.board, boardSizeInt)
                    print("Board is Full!\nIts a Draw! End of Game..")
                    break
                else:
                    self.print_board(self.board, boardSizeInt)
                    currPosStr = input("User {}, please choose the position: ".format("1" if isFirstUser else "2"))
                    currPos = self.to_number(currPosStr)

                    if(currPos <= 0):
                        print("Please provide a numeric value in between 1 and {}".format(totalElements))
                        pass
                    else:
                        if(self.is_pos_occupied(self.board, map_index_matrix[currPos])):
                            print("This position is already occupied. Please provide a different position")
                            self.print_board(self.board, boardSizeInt)
                        else:
                            rowColList = map_index_matrix[currPos]
                            element = " X " if isFirstUser else " O "
                            print()
                            self.board[rowColList[0]][rowColList[1]] = element
                            isFirstUser = False if isFirstUser == True else True

                            if(self.validate_diagonal_rule(self.board, map_index_matrix, boardSizeInt, element) == True):
                                self.print_board(self.board, boardSizeInt)
                                print( element + "is the winner! Won via Diagonal Rule.. End of Game..")
                                break
                            elif(self.validate_vertical_rule(self.board, element, currPos, map_index_matrix, boardSizeInt)):
                                self.print_board(self.board, boardSizeInt)
                                print( element + "is the winner! Won via Vertical Rule. End of Game..")
                                break
                            elif(self.validate_horizontal_rule(self.board, currPos, map_index_matrix, boardSizeInt)):
                                self.print_board(self.board, boardSizeInt)
                                print( element + "is the winner! Won via Horizontal Rule. End of Game..")
                                break
            
    def print_board(self, board, boardSizeInt):
        for i in range(0, boardSizeInt):
            for j in range(0, boardSizeInt):
                if(j == boardSizeInt - 1):
                    print(board[i][j], end="")
                else:
                    print(board[i][j], end="|")
            print()
            if(i < boardSizeInt - 1):
                for j in range(0, boardSizeInt):
                    print(u"\u2015\u2015\u2015", end=" ")    
            print()
        
    def to_number(self, inputVal):
        try:
            return int(inputVal)
        except ValueError:
            return 0
        
    def is_board_full(self, board):
        isBoardFull = True
        for row in board:
            for col in row:
                if col == " * ":
                    isBoardFull = False
                    break
                    
        return isBoardFull
    
    def is_pos_occupied(self, board, rowColList):
        isPosOccupied = True
        if(board[rowColList[0]][rowColList[1]] == " * "):
            isPosOccupied = False
        return isPosOccupied
    
    def validate_diagonal_rule(self, board, map_index_matrix, boardSizeInt, element):
        
        curPos = 1
        isSame = True
        curList = map_index_matrix[curPos]
        curElement = board[curList[0]][curList[1]]
        
        while(isSame == True):
            
            nextPos = curPos + boardSizeInt + 1
            if(nextPos > (boardSizeInt * boardSizeInt)):
                break
            
            nextList = map_index_matrix[nextPos]
            nextElement = board[nextList[0]][nextList[1]]
            
            if(curElement != " * " and nextElement != " * " and curElement == nextElement):
               curPos = nextPos
            else:
                isSame = False
                break
        
        if isSame == False:
            
            isSame = True
            curPos = boardSizeInt
            curList = map_index_matrix[curPos]
            curElement = board[curList[0]][curList[1]]
            
            while(isSame == True):
                nextPos = curPos + boardSizeInt - 1
                
                if(nextPos >= (boardSizeInt * boardSizeInt)):
                    break
                
                nextList = map_index_matrix[nextPos]
                nextElement = board[nextList[0]][nextList[1]]
                
                if(curElement != " * " and nextElement != " * " and curElement == nextElement):
                   curPos = nextPos
                else:
                    isSame = False
                    break
            
        return isSame
    
    def validate_vertical_rule(self, board, element, curPos, map_index_matrix, boardSizeInt):
        
        isSame = True
        curPosTemp = curPos
        
        while(curPosTemp > 0):
            if((curPosTemp - boardSizeInt) <= 0):
                break
                
            curPosTemp = curPosTemp - boardSizeInt
        
        curList = map_index_matrix[curPosTemp]
        curElement = board[curList[0]][curList[1]]
            
        while(isSame == True):
            
            nextPos = curPosTemp + boardSizeInt
            
            if(nextPos > (boardSizeInt * boardSizeInt)):
                break
                
            nextList = map_index_matrix[nextPos]
            nextElement = board[nextList[0]][nextList[1]]
            
            if(curElement != " * " and nextElement != " * " and curElement == nextElement):
               curPosTemp = nextPos
            else:
                isSame = False
                break
                
        return isSame
    
    def validate_horizontal_rule(self, board, currPos, map_index_matrix, boardSizeInt):
        
        currPosTemp = 1
        isSame = True
        
        if currPos > boardSizeInt:
            currPosTemp = currPos
            while(currPosTemp > 0):
                currPosTemp = currPosTemp - 1
                if(currPosTemp % boardSizeInt == 0):
                    break
            currPosTemp = currPosTemp + 1
        
        currList = map_index_matrix[currPosTemp]
        element = board[currList[0]][currList[1]]
        
        for i in range (currPosTemp, currPosTemp + boardSizeInt - 1):
            
            nextPos = i + 1
            
            nextList = map_index_matrix[nextPos]
            nextElement = board[nextList[0]][nextList[1]]
            
            if(element != " * " and nextElement != " * " and element == nextElement):
               curPosTemp = nextPos
            else:
                isSame = False
                break
        
        return isSame
```
